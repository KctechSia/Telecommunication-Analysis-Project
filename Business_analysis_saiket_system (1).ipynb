```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bins = [0, 12, 24, 36, 48, 60, 72],
labels = ['First Year', 'Second Year', 'Third Year', 'Fourth Year', 'Fifth Year', 'Sixth Year']
df['tenure_group'] = pd.cut(df['tenure'], bins = bins, labels = labels)

tenure_stats = df.groupby('tenure_group', observed = True).agg({
    'monthlycharges': 'mean',
    'churn': lambda x: (x == 'Yes').mean() 
}).reset_index()

tenure_stats.columns = ['Tenure Group', 'Avg Monthly Charges', 'Churn Rate']

# Plot churn rate by tenure group
plt.figure(figsize=(8, 5))
sns.lineplot(data=tenure_stats, x='Tenure Group', y='Churn Rate', marker='o')
plt.title('Churn Rate by Customer Lifecycle Stage')
plt.xlabel('Tenure Group (Months)')
plt.ylabel('Churn Rate')
plt.ylim(0, 1)  # because churn rate is between 0 and 1
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
bins = [0, 12, 36, 72]
labels = ['0–12', '13–36', '37+']
df['tenure_segment'] = pd.cut(df['tenure'], bins=bins, labels=labels, right=True)
grouped = df.groupby('tenure_segment', observed=True)['monthlycharges'].mean().reset_index()


# Plotting
plt.figure(figsize=(8,5))
plt.bar(grouped['tenure_segment'], grouped['monthlycharges'], color='skyblue')

# Titles and labels
plt.title('Average Monthly Charges by Tenure Group')
plt.xlabel('Tenure Segment')
plt.ylabel('Average Monthly Charges')
plt.grid(axis='y')
plt.show()

```
